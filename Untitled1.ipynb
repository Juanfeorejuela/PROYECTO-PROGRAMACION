```python
import random

rutinas = {
    "pecho": {
        "fuerza": [
            "Press banca pesado", "Press inclinado barra", "Fondos lastrados",
            "Press declinado", "Press mancuerna pesado"
        ],
        "hipertrofia": [
            "Press banca", "Aperturas mancuerna", "Press inclinado mancuerna",
            "Cruce poleas", "Flexiones lastre ligero"
        ],
        "resistencia": [
            "Press banca ligero", "Flexiones altas repeticiones",
            "Aperturas polea", "Press mancuernas ligero"
        ]
    },
    "espalda": {
        "fuerza": [
            "Peso muerto", "Dominadas lastradas", "Remo barra pesado",
            "Remo T-bar", "Dominadas pronas"
        ],
        "hipertrofia": [
            "Jal√≥n al pecho", "Remo mancuerna", "Remo polea baja",
            "Dominadas", "Pull-over"
        ],
        "resistencia": [
            "Jal√≥n ligero", "Remo polea ligero", "Dominadas asistidas"
        ]
    },
    "piernas": {
        "fuerza": [
            "Sentadilla pesada", "Prensa inclinada", "Peso muerto rumano",
            "Sentadilla frontal", "Zercher squat"
        ],
        "hipertrofia": [
            "Sentadilla", "Zancadas", "Curl femoral",
            "Extensi√≥n de piernas", "Prensa moderada"
        ],
        "resistencia": [
            "Sentadilla ligera", "Prensa ligera", "Extensi√≥n piernas 25 reps"
        ]
    },
    "hombros": {
        "fuerza": [
            "Press militar barra", "Elevaciones laterales pesadas",
            "Press arnold", "Press tras nuca"
        ],
        "hipertrofia": [
            "Press militar", "Elevaciones laterales", "P√°jaros",
            "Press mancuernas", "Face pulls"
        ],
        "resistencia": [
            "Elevaciones ligeras", "Press ligero", "P√°jaros ligero"
        ]
    },
    "brazos": {
        "fuerza": [
            "Curl barra pesado", "Fondos lastrados",
            "Curl martillo pesado", "Press cerrado"
        ],
        "hipertrofia": [
            "Curl mancuerna", "Extensi√≥n tr√≠ceps cuerda",
            "Curl martillo", "Curl concentrado"
        ],
        "resistencia": [
            "Curl ligero", "Extensi√≥n cuerda ligero", "Martillo ligero"
        ]
    }
}

def obtener_series_reps(nivel, objetivo):
    if nivel == "principiante":
        if objetivo == "fuerza": return "3 series de 5-6 repeticiones"
        if objetivo == "hipertrofia": return "3 series de 10-12 repeticiones"
        return "3 series de 15-20 repeticiones"

    if nivel == "intermedio":
        if objetivo == "fuerza": return "4 series de 5 repeticiones"
        if objetivo == "hipertrofia": return "4 series de 10-12 repeticiones"
        return "4 series de 20 repeticiones"

    if nivel == "avanzado":
        if objetivo == "fuerza": return "5 series de 3-5 repeticiones"
        if objetivo == "hipertrofia": return "5 series de 8-12 repeticiones"
        return "5 series de 20-25 repeticiones"


def generar_rutina(grupo, objetivo, nivel):
    if grupo not in rutinas or objetivo not in rutinas[grupo]:
        return ["Grupo u objetivo inv√°lido"]

    ejercicios = random.sample(rutinas[grupo][objetivo], 3)
    sr = obtener_series_reps(nivel, objetivo)

    rutina_final = []
    for e in ejercicios:
        rutina_final.append(f"{e} ‚Äî {sr}")

    return rutina_final


def exportar_txt(nombre, contenido):
    with open(nombre + ".txt", "w", encoding="utf-8") as f:
        f.write(contenido)
    print(f"\nArchivo TXT '{nombre}.txt' creado exitosamente.")


def exportar_pdf(nombre, contenido):
    try:
        from reportlab.platypus import SimpleDocTemplate, Paragraph
        from reportlab.lib.styles import getSampleStyleSheet

        pdf = SimpleDocTemplate(nombre + ".pdf")
        estilos = getSampleStyleSheet()
        story = [Paragraph(contenido.replace("\n", "<br/>"), estilos["Normal"])]
        pdf.build(story)
        print(f"Archivo PDF '{nombre}.pdf' creado exitosamente.")
    except:
        print("No se pudo crear el PDF (falta reportlab).")



def menu():
    print("\n------------------------------------")
    print("     ‚òùÔ∏è ENTRENADOR PERSONAL ‚òùÔ∏è")
    print("------------------------------------")

    dia = input("D√≠a de la semana: ").lower()
    objetivo = input("Objetivo (fuerza / hipertrofia / resistencia): ").lower()
    grupo = input("Grupo muscular (pecho / espalda / piernas / hombros / brazos): ").lower()
    nivel = input("Nivel (principiante / intermedio / avanzado): ").lower()

    rutina = generar_rutina(grupo, objetivo, nivel)

    print("\n====================================")
    print(f" üìÖ Rutina para el d√≠a: {dia.upper()}")
    print("====================================")
    for r in rutina:
        print(f"- {r}")

    texto_exportar = f"Rutina del d√≠a {dia}\nObjetivo: {objetivo}\nNivel: {nivel}\nGrupo muscular: {grupo}\n\nEjercicios:\n"
    for r in rutina:
        texto_exportar += f"- {r}\n"

    print("\n¬øDeseas exportar tu rutina?")
    print("1. S√≠, en TXT")
    print("2. S√≠, en PDF")
    print("3. No")

    opcion = input("Elige: ")

    if opcion == "1":
        exportar_txt("mi_rutina", texto_exportar)
    if opcion == "2":
        exportar_pdf("mi_rutina", texto_exportar)

    print("\n¬°Rutina generada con √©xito! üí™üî•")

menu()
```
